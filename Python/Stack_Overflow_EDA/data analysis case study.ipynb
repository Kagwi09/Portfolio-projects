#### Data Mining and Cleaning
* To Futher analyse this data, we will see how continents are represented in this data

```python
continents = pd.read_html('https://simple.wikipedia.org/wiki/List_of_countries_by_continents')
```

```python
len(continents)
```

* A number of these countries have characters and leading/trailing whitespaces that might result in wrong merges
* A function that removes these characters is required to clean this data

```python
import re
def clean_country_name(country_name):
    # Remove anything inside brackets and the brackets themselves
    cleaned_name = re.sub(r'\[.*?\]', '', country_name) # Remove text in brackets
    # Remove asterisks and any leading/trailing whitespace
    cleaned_name = cleaned_name.replace('*', '').strip()
    return cleaned_name
```

* The re library in Python provides support for working with regular expressions, which are sequences of characters that define search patterns. 
* It's a powerful tool for string manipulation and can be used for tasks such as searching, replacing, and splitting strings. Here’s a breakdown of how the re library works and some common functions you might use:
    * re.sub(pattern, replacement, string): Replaces occurrences of the pattern with the replacement string.

```python
Africa = continents[0]
Africa.to_csv(r"c:\\Users\\MWANGI\\Desktop\\stack-overflow-developer-survey-2023\\Africa.csv")
Overseas_territories = continents[1]
Overseas_territories.to_csv(r"c:\\Users\\MWANGI\\Desktop\\stack-overflow-developer-survey-2023\\Overseas_territories.csv")
Asia = continents[2]
Asia.to_csv(r"c:\\Users\\MWANGI\\Desktop\\stack-overflow-developer-survey-2023\\Asia.csv")
Europe = continents[3]
Europe.to_csv(r"c:\\Users\\MWANGI\\Desktop\\stack-overflow-developer-survey-2023\\Europe.csv")
N_America = continents[4]
N_America.to_csv(r"c:\\Users\\MWANGI\\Desktop\\stack-overflow-developer-survey-2023\\N_America.csv")
S_America = continents[5]
S_America.to_csv(r"c:\\Users\\MWANGI\\Desktop\\stack-overflow-developer-survey-2023\\S_America.csv")
Oceania = continents[6]
Oceania.to_csv(r"c:\\Users\\MWANGI\\Desktop\\stack-overflow-developer-survey-2023\\Oceania.csv")
```

```python
# Clean country names in each continent DataFrame
# Each table has a unique name representing the column with the country name
# Rename the columns for uniformity during merging
Africa['Country'] = Africa['English Names [1][2][3][4]'].apply(clean_country_name)
Overseas_territories['Country'] = Overseas_territories['English names'].apply(clean_country_name)
Asia['Country'] = Asia['English Names [4][10][11][12]'].apply(clean_country_name)
Europe['Country'] = Europe[2].apply(clean_country_name)
N_America['Country'] = N_America['English Names [4][12][19][20]'].apply(clean_country_name)
S_America['Country'] =S_America['English name'].apply(clean_country_name)
Oceania['Country'] = Oceania['English name'].apply(clean_country_name)
```

```python
# Create a dictionary mapping each country to its continent
# The key is the country and the value is the continent
country_to_continent = {}
```

```python
# Adding each continent's countries to the dictionary

for country in Africa['Country']:
    country_to_continent[country] = 'Africa'
for country in Asia['Country']:
    country_to_continent[country] = 'Asia'
for country in N_America['Country']:
    country_to_continent[country] = 'North America'
for country in S_America['Country']:
    country_to_continent[country] = 'South America'
for country in Oceania['Country']:
    country_to_continent[country] = 'Oceania'
for country in Overseas_territories['Country']:
    country_to_continent[country] = 'Overseas territory'
for country in Europe['Country']:
    country_to_continent[country] = 'Europe'
```

```python
# Step 2: Map the dictionary to the 'Country' column in merged_data
merged_data['Continent'] = merged_data['Country'].map(country_to_continent)
```

```python
# Check for NaN values
blank_continent_data = merged_data[merged_data['Continent'].isna()]
blank_continent_data

# fill blank data manually 
merged_data.loc[merged_data['Country'] == 'Bosnia and Herz.', 'Continent'] = 'Europe'
merged_data.loc[merged_data['Country'] == 'Dominican Rep.', 'Continent'] = 'South America'
merged_data.loc[merged_data['Country'] == 'Hong Kong', 'Continent'] = 'Asia'
merged_data.loc[merged_data['Country'] == 'Palestine', 'Continent'] = 'Asia'
merged_data.loc[merged_data['Country'] == 'United States of America', 'Continent'] = 'North America'
```
